```python
rating_with_name=book.merge(rating,on='ISBN')
```

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
num_rating_df.sample(5)
```

```python
avg_rating_df=rating_with_name.groupby(['Book-Title']).mean()['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
avg_rating_df.sample(3)
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df.shape
```

```python
""""We can clearly see only 186 books have rating more than 250 among all the books"""
```

```python
top_popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
top_popular_df.sample(3)
```

```python
"""one book is having multiple different ISBN,It has to be removed,after removing duplicates now we are left with same top 50 books"""
top_popular_df=top_popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

**Collabrative based recommendation system**

```python
temp=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
```

```python
imp_users=temp[temp].index
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(imp_users)]
```

```python
temp=filtered_rating.groupby(['Book-Title']).count()['Book-Rating']>=50
```

```python
famous_books=temp[temp].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating=final_rating.drop_duplicates()
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
"""All the books are now in vector form and to find similarity between books can be calculated by using euclidean distance formula"""
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
'''It gives euclidean distance of one vector with all the other vectors'''
```

```python
similarity_score
```

```python
pd.DataFrame(similarity_score)
```
